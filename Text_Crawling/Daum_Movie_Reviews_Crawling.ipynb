## 영화 링크 수집

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.daum.net/ranking/boxoffice/yearly'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)
else : 
    print(response.status_code)
```

```python
# 남산의 부장들
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_item > div.poster_info > a
```

```python
# 다만 악에서 구하소서
#mainContent > div > div.box_boxoffice > ol > li:nth-child(2) > div > div.thumb_item > div.poster_info > a
```

```python
hrefs = []

for h in soup.select('.poster_info'):
    hrefs.append(h.find("a")["href"])
    
hrefs
```

## 영화 정보 수집

```python
# https://chromedriver.chromium.org/downloads
```

```python
from selenium import webdriver
import time
from tqdm import tqdm
```

```python
# 남산의 부장들
# https://movie.daum.net/moviedb/grade?movieId=122091
```

```python
# 다만 악에서 구하소서
# https://movie.daum.net/moviedb/grade?movieId=127897
```

```python
pre_url = 'https://movie.daum.net/moviedb/grade?movieId='
```

```python
movie_reviews = []

start_time = time.time()
for href in tqdm(hrefs):
    url = pre_url + href.split('=')[-1]
    
    driver = webdriver.Chrome('C:/Program Files/chromedriver/chromedriver.exe')
    driver.implicitly_wait(30)
    driver.get(url)
    
    while True:
        try:
            btn_more = driver.find_element_by_css_selector('div.alex_more')
            btn_more.click()
            time.sleep(3)
        except:
            break
    
    time.sleep(3)
    
    reviews = driver.find_elements_by_css_selector('div.cmt_info')
    temp = []
    for review in reviews:
        temp.append(review.text)
    movie_reviews.append(temp)
    print('Movie ID :', href.split('=')[-1], ', Movie Review Count :', len(reviews))

    driver.quit()
    time.sleep(3)
    
end_time = time.time()

print('Running Time : ', end_time - start_time)
```

## 영화 정보 저장

```python
import pandas as pd
import openpyxl
```

```python
write_wb = openpyxl.Workbook()
write_ws = write_wb.active
```

```python
write_ws = write_wb.active
write_ws['A1'] = 'ID'
write_ws['B1'] = 'DATE'
write_ws['C1'] = 'RATING'
write_ws['D1'] = 'COMMENT'
write_ws['E1'] = 'POSITIVE_NUM'
write_ws['F1'] = 'NEGATIVE_NUM'
```

```python
for movie_review in tqdm(movie_reviews):
    for review in movie_review:
        temp = review.split('\n')
        try:
            write_ws.append([temp[0], temp[1], temp[2], temp[3], temp[5], temp[7]])
            write_wb.save('./daum_movie_reviews.xlsx')
        except:
            #print(review)
            pass
```
